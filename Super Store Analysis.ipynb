```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import squarify
```

```python
# Load dataset
sales_data =pd.read_csv ('Sample - Superstore.csv', encoding = 'latin1')
sales_data.head()
```

```python
sales_data.info
```

```python
sales_data.describe()
```

```python
# Analysis of sales and profit by landmark
# By Region
sales_data_stackb = sales_data[['Region','Sales','Profit']]
sales_data_stackb = sales_data_stackb.groupby(['Region']).sum().reset_index()
plt.figure(figsize=[10,15])
plt.subplot(3,1,1)
plt.bar(x=sales_data_stackb['Region'], height=sales_data_stackb['Sales'] )
plt.bar(x=sales_data_stackb['Region'], height=sales_data_stackb['Profit'], bottom=sales_data_stackb['Sales'], )
plt.title("Sales & Profit By Region", size=20, pad=20)
plt.legend(['Sales','Profit'])
```

```python
# Analysis of sales and profit by landmark
# By State
sales_data_stackb = sales_data[['State','Sales','Profit']]
sales_data_stackb = sales_data_stackb.groupby(['State']).sum().reset_index()
plt.figure(figsize=[25,35])
plt.subplot(3,1,1)
plt.bar(x=sales_data_stackb['State'], height=sales_data_stackb['Sales'] )
plt.bar(x=sales_data_stackb['State'], height=sales_data_stackb['Profit'], bottom=sales_data_stackb['Sales'], )
plt.xticks(rotation=90)
plt.title("Sales & Profit By State", size=20, pad=20)
plt.legend(['Sales','Profit'])
```

```python
# Analysis of Sales by Shipment Mode

sales_data_stackb = sales_data[['Ship Mode','Sales','Profit']]
sales_data_stackb = sales_data_stackb.groupby(['Ship Mode']).sum().reset_index()
plt.figure(figsize=[10,15])
plt.subplot(3,1,1)
plt.bar(x=sales_data_stackb['Ship Mode'], height=sales_data_stackb['Sales'] )
plt.bar(x=sales_data_stackb['Ship Mode'], height=sales_data_stackb['Profit'], bottom=sales_data_stackb['Sales'], )
plt.title("Sales & Profit By Shipment Method", size=20, pad=20)
plt.legend(['Sales','Profit'])
```

```python

# By Segment
sales_data_tree = sales_data[['Segment','Sales']]
sales_data_tree = sales_data_tree.groupby(['Segment']).sum().reset_index()
plt.figure(figsize=[10,8])

squarify.plot(sizes=sales_data_tree['Sales'], label= sales_data_tree['Segment'])
plt.title("Sales Across Different Segment", size=20, pad=20)
plt.axis('off')

plt.show()
```

```python
#Sales and Profit by Category

sales_data_stackb = sales_data[['Category','Sales','Profit']]
sales_data_stackb = sales_data_stackb.groupby(['Category']).sum().reset_index()
plt.figure(figsize=[10,15])
plt.subplot(3,1,1)
plt.bar(x=sales_data_stackb['Category'], height=sales_data_stackb['Sales'])
plt.bar(x=sales_data_stackb['Category'], height=sales_data_stackb['Profit'], bottom=sales_data_stackb['Sales'])
plt.title("Sales & Profit By Category", size=20, pad=20)
plt.legend(['Sales','Profit'])
```

```python
sales_data_stackb = sales_data[['Sub-Category','Sales','Profit']]
sales_data_stackb = sales_data_stackb.groupby(['Sub-Category']).sum().reset_index()
plt.figure(figsize=[10,15])
plt.subplot(3,1,1)
plt.bar(x=sales_data_stackb['Sub-Category'], height=sales_data_stackb['Sales'])
plt.bar(x=sales_data_stackb['Sub-Category'], height=sales_data_stackb['Profit'], bottom=sales_data_stackb['Sales'])
plt.xticks(rotation=90)
plt.title("Sales & Profit By Sub-Category", size=20, pad=20)
plt.legend(['Sales','Profit'])
```
